```python
dummy= pd.get_dummies(object_cols["Type"],drop_first=True,dtype=int,dummy_na=True)
dummy
```

```python
import numpy as np
dummy = dummy.rename(columns={np.nan: 'nan_col'})
# Apply the condition
dummy.loc[dummy['nan_col'] == 1, 'Builder_Floor'] = np.nan
object_cols[["Type"]]=dummy[["Builder_Floor"]]
```

```python
dummy= pd.get_dummies(object_cols["Transaction"],drop_first=True,dtype=int,dummy_na=True)
dummy
```

```python
import numpy as np
dummy = dummy.rename(columns={np.nan: 'nan_col'})
# Apply the condition
dummy.loc[dummy['nan_col'] == 1, 'Resale'] = np.nan
object_cols[["Transaction"]]=dummy[["Resale"]]
```

```python
dummy= pd.get_dummies(object_cols["Furnishing"],drop_first=True,dtype=int,dummy_na=True)
dummy
```

```python
dummy = dummy.rename(columns={np.nan: 'nan_col'})
# Apply the condition
dummy.loc[dummy['nan_col'] == 1, ['Semi-Furnished','Unfurnished']] = np.nan
object_cols[['Semi-Furnished', 'Unfurnished']] = dummy[['Semi-Furnished', 'Unfurnished']]
object_cols = object_cols.drop('Furnishing', axis=1)
```

```python
dummy= pd.get_dummies(object_cols["Status"],drop_first=True,dtype=int,dummy_na=True)
dummy
```

```python
dummy = dummy.rename(columns={np.nan: 'nan_col'})
# Apply the condition
dummy.loc[dummy['nan_col'] == 1, 'Ready_to_move'] = np.nan
object_cols[["Status"]]=dummy[["Ready_to_move"]]
```

```python
object_cols
```

```python
object_cols
```

```python
object_cols.drop("Locality",axis=1,inplace=True)
```

```python
object_cols = object_cols.astype("Int64")
```

```python
object_cols
```

```python
scaler = StandardScaler()
object_cols_sc=scaler.fit_transform(object_cols)
object_cols_sc
```

```python
imputer = KNNImputer(n_neighbors=5)
object_cols_imput=imputer.fit_transform(object_cols_sc)
object_cols_imput
```

```python
data_imputed = scaler.inverse_transform(object_cols_imput)
data_imputed
```

```python
imputed_df = pd.DataFrame(data_imputed,columns=object_cols.columns)
imputed_df
```

```python
object_cols = imputed_df.round().astype("Int64")
object_cols
```

```python
final_dataset=pd.concat([int_cols,float_cols,object_cols],axis=1)
# Impute missing values in the final dataset
imputer = KNNImputer(n_neighbors=5)
final_dataset = imputer.fit_transform(final_dataset)
final_dataset = pd.DataFrame(final_dataset, columns=pd.concat([int_cols,float_cols,object_cols],axis=1).columns)
```

```python
independent=final_dataset.drop("Price",axis=1)
dependent=final_dataset["Price"]
```

```python
from sklearn.model_selection import train_test_split
X_test,X_train,Y_test,Y_train=train_test_split(independent,dependent,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)
r_score
```
