```python
import pandas as pd
```

```python
dataset=pd.read_csv('/content/chennai_rainfall_2019_2023_mm.csv')
```

```python
dataset
```

```python
independent=dataset[["Temperature (°C)","Humidity (%)","Wind Speed (km/h)"]]
dependent=dataset[["Rainfall (mm)"]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=40)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor= RandomForestRegressor()
regressor.fit(x_train,y_train)
```

```python
y_predi=regressor.predict(x_test)
```

```python
y_predi
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_predi)
```
